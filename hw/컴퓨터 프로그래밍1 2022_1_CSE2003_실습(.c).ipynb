```c
#include<stdio.h>

int is_leap(int year);
int near_leap(int year);

int main(void){
    int year=0;
    int result = 0;
    int temp = 0;
    printf("Enter the year to be tested: ");
    scanf("%d", &year);
    result = is_leap(year);
    if(result){
        printf("It is a leap year.\n");
        }
    else{
        printf("It is not a leap year.\n");
        temp = near_leap(year);
        printf("%d is a leap year.\n", temp);
        }
    }
    





int is_leap(int year){
    int a = 0;
    int b = 0;
    int c = 0;
    int d = 0;
    a = (year %4);
    b = (year %100);
    c = (year %400);
    if (((!a) && (b)) || (!(c)) ) {   // 4로 나누어 떨어지면 참이므로 !a, 100으로 나누어 떨어지지 않으면 참이므로 b, 400으로 나누어 떨어지면 참이므로 !c
        d = 1;
    ;}
    else{
        d=0;
         } 
    return d;
}

int near_leap(int year){   
    
    if(is_leap(--year)){  //이전 년도 기준으로 7년 안에 윤년이 나오므로 7번만 반속해서 윤년인지 알아본다. 
        return year;
    
        }
    else if(is_leap(--year)){
        return year;}
    else if(is_leap(--year)){
        return year;}
    else if(is_leap(--year)){
        return year;}
    else if(is_leap(--year)){
        return year;}
    else if(is_leap(--year)){
        return year;}
    else if(is_leap(--year)){
        return year;}
}
```

```c
#include<stdio.h>

int fib(int n){
    int now=0, a=0, b=1, next=0, count= -1;
    // 일단 실행이기 때문에, 0번째값은 한번 실행되고 난 뒤의 값으로, -1부터 시작
    // n이 8이면 다음과 같은 순서로 진행됨.
    // now=0, a=1, b=1, next=1 count0 n=8
    // now=1, a=1, b=2, next=2 count1 n=8
    // now=1, a=2, b=3, next=3 count2 n=8
    // now=2, a=3, b=5, next=5 count3 n=8
    // now=3, a=5, b=8, next=8 count4 n=8
    // now=5, a=8, b=13, next=13 count5 n=8
    // now=8, a=13, b=21, next=21 count6 n=8
    // now=13, a=21, b=34, next=34 count7 n=8
    // now=21, a=34, b=55, next=55 count8 n=8
    do {
        now = a;
        next = a+b;
        a=b;
        b=next;
        count++;
        //printf("now=%d, a=%d, b=%d, next=%d count%d n=%d\n",now, a, b, next, count, n);
    }while(count < n);
    return now;
    
}



int main(void){
    int n=0, f=0; //n, n번쨰 피보나치를 받아올 변수f.  
    printf("input n: ");
    scanf("%d", &n);
    if (n<0){
        printf("n cannot be negative number");
    }
    else{
        f = fib(n);
        printf("fib(%d) = %d",n, f);
    }
    
    
}
```

```c
#include<stdio.h>



void swap(int*, int*);
int sumof(int, int);


int main(void){
    int num1=0, num2=0, sum=0;
    printf("Input first number : ");
    scanf("%d", &num1);
    printf("input second number : ");
    scanf("%d", &num2);

    if(num1<1 || num2<1){
        printf("Input is less than 1\n");
        }
    else {
        if(num1>num2){
            swap(&num1, &num2);
            }
        sum=sumof(num1, num2);
        printf("Result: %d\n", sum);
    }
}














void swap(int* num1, int* num2){
    int temp=0;
    temp = *num1;
    *num1 = *num2;
    *num2 = temp;
    }

int sumof(int num1, int num2){
    int sum = 0;
    for(num1; num1<=num2; num1++){
        if(!(num1%2) || !(num1%3)){
            sum = sum+num1;
            //printf("%d %d\n", sum, num1);
            }
        }
    return sum;
    }


```
